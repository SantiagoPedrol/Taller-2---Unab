```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# Plotly Libraris
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

```python
data_set = pd.read_csv('/content/drive/MyDrive/heart_2022_no_nans.csv')
```

```python
smoker_status = data_set.SmokerStatus
print(smoker_status)
```

```python
# Cargar el dataset

# Ver las primeras filas del dataset
#print(heart_2022.head())

# Calcular estadísticas descriptivas
#print(heart_2022.describe())

sns.countplot(data=data_set, x='Sex')
plt.title('Distribución del Sexo')
plt.xlabel('Sexo')
plt.ylabel('Cantidad')
plt.show()
```

```python
# Visualizar la distribución de fumadores y no fumadores
plt.figure(figsize=(6, 4))
sns.countplot(data=data_set, x='SmokerStatus')
plt.title('Distribución de fumadores y no fumadores')
plt.xlabel('Estado del fumador')
plt.ylabel('Cantidad')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# Comparar distribución de género entre fumadores y no fumadores
plt.figure(figsize=(8, 6))
sns.countplot(data=data_set, x='Sex', hue='SmokerStatus')
plt.title('Distribución de Género por Estado de Fumador')
plt.xlabel('Género')
plt.ylabel('Cantidad')
plt.legend(title='Estado de Fumador')
plt.show()
```

```python
# Comparar actividad física
plt.figure(figsize=(10, 6))
sns.countplot(data=data_set, x='PhysicalActivities', hue='SmokerStatus')
plt.title('Actividad Física por Estado de Fumador')
plt.xlabel('Actividad Física')
plt.ylabel('Cantidad')
plt.legend(title='Estado de Fumador')
plt.show()
```

```python
# Comparar consumo de alcohol con el orden del eje x invertido
plt.figure(figsize=(10, 6))
sns.countplot(data=data_set, x='AlcoholDrinkers', hue='SmokerStatus', order=['Yes', 'No'])
plt.title('Consumo de Alcohol por Estado de Fumador')
plt.xlabel('Consumo de Alcohol')
plt.ylabel('Cantidad')
plt.legend(title='Estado de Fumador')
plt.show()
```

```python
# Crear boxplot para comparar la edad entre fumadores y no fumadores
plt.figure(figsize=(8, 6))
sns.boxplot(data=data_set, x='SmokerStatus', y='AgeCategory')
plt.title('Distribución de Edad por Estado de Fumador')
plt.xlabel('Estado de Fumador')
plt.ylabel('Categoría de Edad')
plt.xticks(rotation=45, ha='right')
plt.show()
```
